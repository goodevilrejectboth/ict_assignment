```python
 # 1  FizzBuzz: For a given range of numbers (e.g., 1 to 100), print "Fizz" if the number is divisible by 3, "Buzz" if it's divisible by 5, and "FizzBuzz" if it's divisible by both. Otherwise, print the number itself.
for i in range(1,101):
  if i % 3 == 0 and i % 5 == 0:
    print("FizzBuzz")
  elif i % 3 == 0:
    print('fizz')
  elif i % 5 == 0:
    print('Buzz')
  else:
    print(i)
```
