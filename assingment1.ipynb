```python
# 9
length = float(input('Enter the length of the wall in meters: '))
height = float(input('Enter the height of the wall in meters: '))
cost_per_sq_meter = 25
wall_area = length * height
window_area_single = 1.5 * 1.2
windows_total_area = 2 * window_area_single
paintable_area = wall_area - windows_total_area
total_cost = paintable_area * cost_per_sq_meter
print(f"Total wall area: {wall_area:.2f} sq meters")
print(f"Total window area: {windows_total_area:.2f} sq meters")
print(f"Area to be painted: {paintable_area:.2f} sq meters")
print(f"Total cost to paint: ₹{total_cost:.2f}")
```
