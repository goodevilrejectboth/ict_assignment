```python
# 10. Log File Analyze
records = [
    "INFO Alice Login",
    "ERRO#R Bob Failed",
    "INFO Alice Logout",
    "WARNING Alice Expire",
    "ERROR Bob Failed",
    "INFO Charlie Login"
]

level_count = {}
user_count = {}
error_count = {}

# Process records
for r in records:
    words = r.split(" ")

    level = words[0]
    user = words[1]
    action = words[2]

    # Count levels
    if level in level_count:
        level_count[level] = level_count[level] + 1
    else:
        level_count[level] = 1

    # Count users
    if user in user_count:
        user_count[user] = user_count[user] + 1
    else:
        user_count[user] = 1

    # Count errors
    if level == "ERROR":
        if action in error_count:
            error_count[action] = error_count[action] + 1
        else:
            error_count[action] = 1

# Menu
while True:
    print("\n--- Menu ---")
    print("1. Show level count")
    print("2. Show error count")
    print("3. Show user count")
    print("4. Filter by level")
    print("5. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        print("\nLevel Count")
        for k in level_count:
            print(k, ":", level_count[k])

    elif choice == "2":
        print("\nError Count")
        for k in error_count:
            print(k, ":", error_count[k])

    elif choice == "3":
        print("\nUser Count")
        for k in user_count:
            print(k, ":", user_count[k])

    elif choice == "4":
        lvl = input("Enter level (INFO/WARNING/ERROR): ")
        print("\nRecords:")
        for r in records:
            if r.startswith(lvl):
                print(r)

    elif choice == "5":
        print("Program finished")
        break

    else:
        print("Wrong choice")
```

```python
# Question bank (nested structure)
quiz = {
    "easy": [
        ["is trump funny = ?", "yes"],
        ["what do you think about ram price increase in this days?", "it sucks"]
    ],
    "medium": [
        ["is earth flat = ?", "yes"],
        ["is bat man black or white ?", "a man "]
    ],
    "hard": [
        ["why do humans fuck up mentaly", " we are humans"],
        ["1man+1woman=?", "babies"]
    ]
}

score = 0
correct = 0
wrong = 0

# Fixed order (pseudo-random)
order = ["medium", "easy", "hard"]

print("---- Quiz Started ----")

for level in order:
    print("\nDifficulty:", level)

    for q in quiz[level]:
        question = q[0]
        answer = q[1]

        user_answer = input(question + " ").lower()

        if user_answer == answer:
            print("Correct")

            if level == "easy":
                score = score + 1
            elif level == "medium":
                score = score + 2
            else:
                score = score + 3

            correct = correct + 1
        else:
            print("Wrong")

            if level == "easy":
                score = score - 0
            elif level == "medium":
                score = score - 1
            else:
                score = score - 2

            wrong = wrong + 1

# Final Report
print("\n---- Score Report ----")
print("Correct Answers:", correct)
print("Wrong Answers:", wrong)
print("Final Score:", score)
```
